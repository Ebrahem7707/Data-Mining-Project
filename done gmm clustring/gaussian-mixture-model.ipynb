```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from pandas import DataFrame
# from sklearn import datasets
from sklearn.mixture import GaussianMixture

# Load the iris dataset
data_pd = pd.read_csv("iris.csv")

# Select the first two columnsl
X = data_pd.iloc[:, :2]

# Turn it into a dataframe
d = pd.DataFrame(X)

# Plot the data
plt.scatter(d.iloc[:, 0], d.iloc[:, 1])
plt.show()
```

```python
# Fit the GMM model for the dataset 
# which expresses the dataset as a mixture of 3 Gaussian Distribution
gmm = GaussianMixture(n_components = 3)
gmm.fit(d)

# Assign a label to each sample
labels = gmm.predict(d)
d['labels']= labels
d0 = d[d['labels']== 0]
d1 = d[d['labels']== 1]
d2 = d[d['labels']== 2]

# Get the column names
col_names = d.columns.tolist()

# Plot three clusters in same plot
plt.scatter(d0[col_names[0]], d0[col_names[1]], c ='r')
plt.scatter(d1[col_names[0]], d1[col_names[1]], c ='yellow')
plt.scatter(d2[col_names[0]], d2[col_names[1]], c ='g')
plt.show()
```

```python
# print the converged log-likelihood value
print(gmm.lower_bound_)

# print the number of iterations needed
# for the log-likelihood value to converge
print(gmm.n_iter_)
```
